```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('chunk/chunk0.csv',
                names=['Transaction unique identifier',
                                 'Price', 'Date of Transfer', 'Postcode',
                                 'Property Type', 'Old/New', 'Duration',
                                 'PAON', 'SAON', 'Street', 'Locality',
                                 'Town/City', 'District','Country',
                                 'PPD Category Type','Record Status-monthly file only'])
df.head(3)
```

```python
df['Uniq Adress'] = df['Street'] + ' ' + df['PAON'] + ' ' +df['Locality']
df.head(3)
```

```python
full_price_df = df.groupby('Uniq Adress').sum(['Price'])
print('Table 1')
print(full_price_df)
```

```python
df_1 = df.filter(['Uniq Adress', 'Price'], axis=1)



resault = full_price_df.merge(df_1[['Uniq Adress', 'Price']], how='outer',
                             on='Uniq Adress')


resault['compare'] = resault['Price_x'] != resault['Price_y']
resault.head(5)
```

```python


answer = resault.loc[resault['compare'] == True]
```

```python
a = answer.dropna(how='any')
```

```python
b = a.drop_duplicates(subset=['Uniq Adress'])
```

```python
b['Uniq Adress'].to_csv('b.csv')
```
